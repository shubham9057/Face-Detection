```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv.imread('test.jpg')
img_rgb=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
plt.imshow(img_gray,cmap='gray')
```

```python
model=cv.CascadeClassifier("haar_frontal.xml")
faces=model.detectMultiScale(img_gray,1.3,5)
```

```python
org=img_rgb.copy()
for face in faces:
    a,b,w,h=face
    cv.rectangle(org,(a,b),(a+w,b+h),(255,255,255),5)
plt.imshow(org)
```

```python
# face detection using camera

model=cv.CascadeClassifier("haar_frontal.xml")
cam=cv.VideoCapture(0)
while True:
    _,frame=cam.read()
    fr=cv.flip(frame,+1)
    img_gray=cv.cvtColor(fr,cv.COLOR_BGR2GRAY)
    
    #cv.imshow("face detection",img_gray)
    try:  
        faces=model.detectMultiScale(img_gray,1.3,5)
        for face in faces:
            a,b,w,h=face
            cv.rectangle(fr,(a,b),(a+w,b+h),(255,255,255),5)
    except:
        print("Error")
    cv.imshow("face detection",fr)
    
    key=cv.waitKey(1)
    if key==ord('q'):
        cv.destroyAllWindows()
        break
cam.release()
    
    
```
